```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import os
os.chdir('/Users/emanu/Documents/EBAC')
```

```python
data = pd.read_csv('kc_house_data.csv')
data
```

```python
data['Intercepto'] = 1
data = data[['Intercepto', 'price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']]
print(data.shape)
data.head()
```

```python
Xdata = data[['Intercepto', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']].values
Ydata = data[['price']].values
```

```python
# Dividir en bases de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Xdata, Ydata, test_size=0.3, random_state=1)
```

```python
X = X_train
Y = Y_train
```

```python
# Opcion de formateo numerico 
np.set_printoptions(formatter={'float_kind':'{:f}'.format})
```

```python
XT_X = np.matmul(np.matrix.transpose(X), X)
XT_X
```

```python
XT_X_inv =np.linalg.inv(XT_X)
XT_X_inv
```

```python
XT_Y = np.matmul(np.matrix.transpose(X), Y)
XT_Y
```

```python
betas = np.matmul(XT_X_inv, XT_Y)
betas
```

```python
Y_pred = np.matmul(X_test, betas)
Y_pred
```

```python
# Calculo de residuales
Resid  = Y_test - Y_pred
Resid
```

```python
# Calculo del TSS (Suma Total de Cuadrados)
TSS = np.matmul(np.matrix.transpose(Y), Y) - len(Y)*(Y.mean()**2)
TSS
```

```python
# Calculo del ESS (Suma Explicada de Cuadrados)
ESS = sum((Y - Y.mean())**2)
ESS
```

```python
# Calculo de RSS (Residuales al Cuadrado)
RSS = TSS - ESS
RSS
```

```python
# Calculo del Coeficioente de Determinacion R Cuadrada
RSq = 1 - RSS / TSS
RSq
```

```python
# Calculo del Coeficiente de Determinacion R Cuadrada Ajustada)
RSqA = 1 - (RSS / (X.shape[0] - X.shape[1])) / (TSS / (X.shape[0] -1))
RSqA
```

```python
# Calculo de la varianza del error de regresion
s_cuad = RSS / (len(Y) - X.shape[1])
s_cuad
```

```python
# Desviacion estandar del error de regresion
import math 
s = math.sqrt(s_cuad)
s
```

```python
# Calculo de las t's estadisticas para cada coeficiente de regresion 
result_t = []
for i in range(0, X.shape[1]):
    t = float(betas[i] / (s*math.sqrt(XT_X_inv[i][i])))
    result_t.append(t)
result_t
```

# Criterio 1:

```python
# Obtener valor critico de la t de Student de tablas
import scipy.stats 

grados_libertad = len(Y) - X.shape[1]
# la t critica se obtendra a un nivel de confianza del 95% (Alfa = 5%)
t_critico = abs(scipy.stats.t.ppf(q=0.025, df=grados_libertad))
t_critico
```

```python
for i in range(0, X.shape[1]):
    if (abs(result_t[i]) > t_critico):
        print('Beta', i, 'es significativa') # Aqui se rechaza la hipotesis nula
    else: 
         print('Beta', i, ' NO es significativa') # Aqui NO se rechaza la H0
```

# Criterio 2:

```python
# Calculo de valores p
for i in range(0, X.shape[1]):
    print('Valor p de Beta', i, ':', scipy.stats.t.sf(abs(result_t[i]), df=grados_libertad)*2)
```

# Criterio 3:

```python
# Calculo de intervalos de confianza del 95% para el verdadero valor del coeficiente de cada Beta
for i in range(0, X.shape[1]):
    print('El valor de Beta', i, 'se encuentra entre', float(betas[i]) - t_critico*s*math.sqrt(XT_X_inv[i][i]), 
          'y', float(betas[i]) + t_critico*s*math.sqrt(XT_X_inv[i][i]))
```
